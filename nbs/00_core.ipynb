```python
# export

# Code adapted from: https://gist.github.com/kylebgorman/1081951/bce3de986e4b05fc0b63d4d9e0cfa4bde6664365
def _zeros(*shape):
    """ like this syntax better...a la MATLAB """
    return zeros(shape)

def levenshtein_distance(to_edit, ref, normalized=False, delete=1, insert=1, sub=1):
    D = _zeros(len(to_edit) + 1, len(ref) + 1)
    for i in range(len(to_edit)): 
        D[i + 1][0] = D[i][0] + delete
    for j in range(len(ref)): 
        D[0][j + 1] = D[0][j] + insert
    for i in range(len(to_edit)): # fill out middle of matrix
        for j in range(len(ref)):
            if to_edit[i] == ref[j]:
                D[i + 1][j + 1] = D[i][j] # aka, it's free. 
            else:
                D[i + 1][j + 1] = min(D[i + 1][j] + insert,
                                      D[i][j + 1] + delete,
                                      D[i][j]     + sub)
    return D
```
